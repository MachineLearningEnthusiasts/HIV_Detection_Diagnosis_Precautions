```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
# Path of the file to read
insurance_filepath = "C:\\Users\\SOUMYADIP MALLICK\\OneDrive\\Desktop\\HIV & AIDS\\HIV in child-.csv"

# Read the file into a variable insurance_data
insurance_data = pd.read_csv("C:\\Users\\SOUMYADIP MALLICK\\OneDrive\\Desktop\\HIV & AIDS\\HIV in child-.csv")
```

```python
plt.rcParams["figure.figsize"] = (20,5)
sns.scatterplot(x=insurance_data['Year'], y=insurance_data['value'])
```

```python
plt.rcParams["figure.figsize"] = (20,5)
sns.regplot(x=insurance_data['Year'], y=insurance_data['value'])
```

```python
plt.rcParams["figure.figsize"] = (20,5)
sns.scatterplot(x=insurance_data['Year'], y=insurance_data['value'], hue=insurance_data['Sex'])
```

```python
plt.rcParams["figure.figsize"] = (30,5)

sns.lmplot(x="Year", y="value", hue="Sex", data=insurance_data)
```

```python
plt.rcParams["figure.figsize"] = (20,5)

sns.swarmplot(x=insurance_data['Year'],
              y=insurance_data['value'])
```
