```python
# Import the modules
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv("C:\\Users\\SOUMYADIP MALLICK\\OneDrive\Desktop\\HIV & AIDS\\15 to 49 HIV people.csv")
df
```

```python
names = df['Country'].values
x = np.arange(len(names))
w = 0.3
plt.bar(x-w, df['lower limit'].values, width=w, label='lower limit')
plt.bar(x, df['estimate'].values, width=w, label='estimate')
plt.bar(x+w, df['upper limit'].values, width=w, label='upper limit')
plt.rcParams["figure.figsize"] = (60,20)
plt.xticks(x, names)
plt.ylim([0,25])
plt.tight_layout()
plt.xlabel('country')
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1), fancybox=True, ncol=5)
plt.show()
```

```python

# Set the figure size
plt.rcParams["figure.figsize"] = [7.00, 3.50]
plt.rcParams["figure.autolayout"] = True

# Make a list of columns
columns = ['estimate', 'lower limit', 'upper limit']

# Read a CSV file
df = pd.read_csv("C:\\Users\\SOUMYADIP MALLICK\\OneDrive\Desktop\\HIV & AIDS\\15 to 49 HIV people.csv", usecols=columns)
plt.rcParams["figure.figsize"] = (20,10)
# Plot the lines
df.plot()

plt.show()
```
