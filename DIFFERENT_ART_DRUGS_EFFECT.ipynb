```python
# Import the modules
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data= pd.read_csv("C:\\Users\\SOUMYADIP MALLICK\\OneDrive\\Desktop\\HIV & AIDS\HIV-DRUG.csv")
data
```

```python
drug=data['Drug Name']
eff=data['Effectivity']
x=list(drug)
y=list(eff)
plt.plot(x, y, color = 'g', linestyle = 'dashed', marker = 'o', label = "ART Drugs Effectivity")
plt.xlabel('Drugs')
plt.rcParams["figure.figsize"] = (50,50)
plt.xticks(rotation = 25)
plt.ylabel('Effectivity')
plt.title('Effectivity of ART drugs in HIV treatment')
plt.legend()
plt.show()
```

```python
st_name=data['Drug Name']
marks=data['Effectivity']
x=list(st_name)
y=list(marks)
plt.rcParams["figure.figsize"] = (50,30)
plt.pie(y, labels=st_name, autopct='%.2f%%')
plt.xlabel('Drug Name')
plt.ylabel('Effectivity')
plt.title('Effectivity of ART drugs in HIV treatment')
plt.legend()
plt.show()
```

```python
st_name=data['Drug Name']
marks=data['Effectivity']
x=list(st_name)
y=list(marks)
plt.rcParams["figure.figsize"] = (60,25)
plt.bar(x, y, color = 'g', width = 0.5, label = "Marks")
plt.xlabel('Drug Name')
plt.ylabel('Effectivity')
plt.title('Effectivity of ART drugs in HIV treatmen')
plt.legend()
plt.show()
```
